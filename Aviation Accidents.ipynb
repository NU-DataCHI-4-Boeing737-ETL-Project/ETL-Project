```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/AviationData.csv"
accidents = pd.read_csv(csv_file, encoding='iso8859')
accidents.head()
```

```python
#accidents.to_csv(r'accidents.csv')
```

```python
Adrop_new = accidents[["Event.Id", "Investigation.Type", "Event.Date", "Country", "Airport.Code", "Injury.Severity", "Aircraft.Damage"]]
```

```python
Adrop_new.dropna()
Adrop_new.head()
```

```python
Adrop_new = Adrop_new.loc[Adrop_new['Country'] == "United States", :]
Adrop_new.head()
```

```python
Adrop_new.rename(columns={"Event.Id": "Event_Id", "Investigation.Type" : "Investigation_Type",
                         "Event.Date": "Event_Date", "Airport.Code" : "Airport_Code", "Injurey.Severity": "Injurey_ Severity", 
                        "Aircraft.Damage": "Aircraft_Damage"})
Adrop_new.head()
```

```python
Adrop_new.to_csv(r'Adrop_new.csv')
```
